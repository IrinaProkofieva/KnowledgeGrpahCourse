```python
def AddCharacterIndividual(novel, id, is_real, sex_ch, prof_ch=None, idea_ch=None, cod=None, mention_ch=None):
  id = id.replace(' ', '_')
  individual = URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + id)
  if is_real:
    if idea_ch is None:
      g.add((individual, RDF.type, Acting_character))
    else:
      g.add((individual, RDF.type, Character_with_idea))
      g.add((individual, idea, Literal(idea_ch)))
  else:
    g.add((individual, RDF.type, Mentioned_image))
    if mention_ch is not None:
      m_ch = URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + mention_ch.replace(' ', '_'))
      g.add((m_ch, mention, individual))
  g.add((individual, described_in, novel))
  g.add((individual, sex, Literal(sex_ch)))
  if prof_ch is not None:
    g.add((individual, profession, Literal(prof_ch)))
  if cod is not None:
    g.add((individual, cause_of_death, Literal(cod)))
  return individual

def AddRealPersonIndividual(id, s, prof, birth, death, death_cause=None, prototype_of=None):
  id = id.replace(' ', '_')
  individual = URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + id)
  g.add((individual, RDF.type, Real_human))
  if (s is not None):
    g.add((individual, sex, Literal(s)))
  if prof is not None:
    g.add((individual, profession, Literal(prof)))
  if (birth is not None):
    g.add((individual, year_of_birth, Literal(birth, datatype=XSD.unsignedInt)))
  if (death is not None):
    g.add((individual, year_of_death, Literal(death, datatype=XSD.unsignedInt)))
  if (death_cause is not None):
    g.add((individual, cause_of_death, Literal(death_cause)))
  if (prototype_of is not None):
    chars = prototype_of.split(';')
    for ch in chars:
      prot_of = URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + ch.replace(' ', '_'))
      g.add((individual, is_prototype_of, prot_of))
  return individual

def AddNovel(id, novel_author, year):
  individual = URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + id.replace(' ', '_'))
  g.add((individual, name, Literal(id)))
  g.add((individual, RDF.type, Novel))
  g.add((novel_author, author, individual))
  g.add((individual, year_of_publishing, Literal(year, datatype=XSD.unsignedInt)))
  return individual

def AddRelation(subj, pred, obj):
  if not isinstance(subj, URIRef):
    subj = URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + subj.replace(' ', '_'))
  if not isinstance(obj, URIRef):
    objs = obj.split(';')
    for o in objs:
      obj = URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + o.replace(' ', '_'))
      g.add((subj, pred, obj))
  else:
    g.add((subj, pred, obj))

def AddMovie(id, name_m, countries, genres, rate, year, book):
  individual = URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + str(id) + "_" + name_m.replace(' ', '_'))
  g.add((individual, RDF.type, Movie))
  g.add((individual, name, Literal(name_m)))
  cs = countries.split(';')
  for c in cs:
    cntr = URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + c.replace(' ', '_'))
    g.add((individual, created_in, cntr))
  gs = genres.split(';')
  for gr in gs:
    g.add((individual, genre, Literal(gr)))
  if rate is not None:
    g.add((individual, rating, Literal(rate, datatype=XSD.decimal)))
  g.add((individual, year_of_publishing, Literal(year, datatype=XSD.unsignedInt)))
  g.add((individual, plotted_by, book))
  return individual

def AddCountry(id):
  individual = URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + id.replace(' ', '_'))
  g.add((individual, RDF.type, Country))

def AddActor(id, ch_name, movie_id):
  individual = URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + id.replace(' ', '_'))
  g.add((individual, RDF.type, Real_human))
  g.add((individual, profession, Literal('актер')))
  if ch_name is not None:
    g.add((individual, played_character, URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + ch_name.replace(' ', '_'))))
  g.add((individual, played_in, URIRef("http://www.semanticweb.org/irina/ontologies/2023/10/dostoevskiy#" + movie_id)))
```

## Парсинг и ввод информации о фильмах и актерах

**Источник информации о фильмах и актерах:** https://kinopoiskapiunofficial.tech/documentation/api/#/films/get_api_v2_2_films

Получаем через API информацию о фильмах, снятых на основе романов Достоевского:

```python
import urllib.parse
import urllib.request
import json


names = [
    ('Идиот', Idiot),
    ('Карамазовы', Brothers_Karamazov),
    ('Подросток', Raw_Youth),
    ('Преступление и наказание', Crime_and_Punishment),
    ('Бесы', Demons)
]
films = []

for f_name, f in names:
  url = 'https://kinopoiskapiunofficial.tech/api/v2.2/films?order=RATING&type=ALL&ratingFrom=0&ratingTo=10&yearFrom=1000&yearTo=3000&keyword='+str(f_name.encode()).replace('\\x', '%').replace(' ','%20')[2:-1]+'&page=1'
  headers= {
        'X-API-KEY': 'insert_your_token', #здесь используется токен, полученный при регистрации на сайте
        'Content-Type': 'application/json',
  }

  req = urllib.request.Request(url, headers=headers)
  with urllib.request.urlopen(req) as response:
    response_str = response.read().decode('UTF8')
    response = json.loads(response_str)
    for film in response['items']:
      films.append((film['kinopoiskId'], film['nameRu'], film['countries'], film['genres'], film['ratingKinopoisk'], film['year'], f))
```

```python
# в список фильмов попали лишние - надо убрать
unrelevant = [737716, 423210, 55509, 161214, 316101, 22599, 10723, 490337, 41345, 1253988, 860562, 756288, 408531, 9755, 54350, 21468, 4920653, 3926, 461095, 57421, 121084, 178052, 4401010, 200145, 200187, 511475, 61228, 1252552, 764001, 24140, 100626, 150421, 38895, 413712, 471377, 26755, 927756, 561918, 18912, 395207, 675185, 4367657, 262480, 300061, 371962, 467138, 798111, 693004, 26122, 1009182]
cleared_films = list(filter(lambda f: f[0] not in unrelevant, films))
```

Парсим информацию об актерах, снимавшихся в фильмах, и сопоставляем их роли с ерсонажами книг:

```python
actors = []

for film in cleared_films:
  url = 'https://kinopoiskapiunofficial.tech/api/v1/staff?filmId=' + str(film[0])
  headers= {
        'X-API-KEY': 'insert_your_token', #тут используется токен, полученный при регистрации на сайте
        'Content-Type': 'application/json',
  }

  req = urllib.request.Request(url, headers=headers)
  with urllib.request.urlopen(req) as response:
    response_str = response.read().decode('UTF8')
    response = json.loads(response_str)
    for actor in response:
      if (actor['professionKey'] == 'ACTOR'):
        actors.append((actor['staffId'], actor['nameRu'], actor['nameEn'], actor['description'], str(film[0])+'_'+film[1].replace(' ', '_'), film[6]))
```

Чтобы понять, кого играл актер, ищем самую похожую строку среди персонажей романа - основы фильма для транслитерированной строки с описанием роли:

```python
import difflib
from transliterate import translit, get_available_language_codes

role_to_character = {}

roles = set()
for actor in actors:
  if actor[3] is not None:
    roles.add((actor[3], actor[5]))

cs = {}
cs[Demons] = [dc[0] for dc in demons_chars]
cs[Brothers_Karamazov] = [dc[0] for dc in karamasovs_chars]
cs[Idiot] = [dc[0] for dc in idiot_chars]
cs[Raw_Youth] = [dc[0] for dc in teenager_chars]
cs[Crime_and_Punishment] = [dc[0] for dc in crime_chars]

for r, f in roles:
  clr = difflib.get_close_matches(translit(r, 'ru'), cs[f])
  if len(clr) > 0:
    role_to_character[r] = clr[0]
```
